## Prediction using Decision Tree Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
data = load_iris()
i = pd.DataFrame(data.data,columns=data.feature_names)
print(i.head(5))
o=data.target
```

```python
i.describe()
```

```python
from sklearn.model_selection import train_test_split
i_train,i_test,o_train,o_test=train_test_split(i, o, test_size=0.3, random_state=17)
```

```python
classifier = tree.DecisionTreeClassifier(random_state=15)
classifier = classifier.fit(i_train, o_train)
```

```python
figure = plt.figure(figsize=(15,10))
dt= tree.plot_tree(classifier,feature_names=data.feature_names,class_names=data.target_names,filled=True)
```
