```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
df['Cost'] = df['Sales']-df['Profit']
df['Cost'].head()
```

```python
df['Profit %'] = (df['Profit']/df['Cost'])*100
df
```

```python
df.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

```python
plt.figure(figsize=(16,10))
sns.set_style("darkgrid")
sns.countplot(x='Sub-Category',data=df)
```

```python
sns.countplot(x='Region',data=df)
```

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
statewise=df.groupby(['State'])['Sales'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,10))
plt.title('State vs Sales')
statewise.plot.bar()
```

### From the above Graph we can conclude that California has highest number of sales.

```python
statewise=df.groupby(['State'])['Profit'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,10))
plt.title("State vs Profit")
statewise.plot.bar()
```

### we can conclude from above graph that New York has the 2nd highest sales.
